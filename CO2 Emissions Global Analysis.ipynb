```python
sns.set_theme(style="whitegrid")

# Group the data by 'Year' and sum the 'Kilotons of CO2' to get global emissions per year
global_trend = data.groupby('Year')['Kilotons of Co2'].sum()

# Plot the global CO2 emissions trend over time
plt.figure(figsize=(12, 6))
plt.plot(global_trend.index, global_trend.values, marker='o', color='teal')
plt.title("Global CO2 Emissions Trend (1990-2019)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Total CO2 Emissions (Kilotons)", fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python

# Group the data by 'Region' and sum the 'Kilotons of CO2'
regional_emissions = data.groupby('Region')['Kilotons of Co2'].sum().sort_values(ascending=False)

# Plot the total CO2 emissions by region
plt.figure(figsize=(14, 7))
sns.barplot(x=regional_emissions.index, y=regional_emissions.values, palette='viridis')
plt.title("Total CO2 Emissions by Region (1990-2019)", fontsize=16)
plt.xlabel("Region", fontsize=14)
plt.ylabel("Total CO2 Emissions (Kilotons)", fontsize=14)
plt.xticks(rotation=45)
plt.show()

```

```python
# Trend analysis of CO2 emissions by region over time
plt.figure(figsize=(14, 7))
sns.lineplot(data=data, x='Year', y='Kilotons of Co2', hue='Region', marker='o')
plt.title("CO2 Emissions Trends by Region (1990-2019)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("CO2 Emissions (Kilotons)", fontsize=14)
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Group the data by 'Country' to calculate total CO2 emissions and average CO2 per capita
country_emissions = data.groupby('Country').agg(
    total_co2_emissions=('Kilotons of Co2', 'sum'),
    avg_co2_per_capita=('Metric Tons Per Capita', 'mean')
).reset_index()

# Calculate the correlation between total CO2 emissions and average CO2 per capita
correlation = country_emissions[['total_co2_emissions', 'avg_co2_per_capita']].corr().iloc[0, 1]

print(f"Correlation between total CO2 emissions and average CO2 per capita: {correlation:.2f}")
```

```python
# Scatter plot with regression line and log scale for better spread
plt.figure(figsize=(14, 7))
sns.regplot(
    data=country_emissions,
    x='total_co2_emissions',
    y='avg_co2_per_capita',
    color='darkblue',
    marker='o',
    scatter_kws={'s': 80, 'alpha': 0.7}
)

# Set the x-axis and y-axis to a log scale
plt.xscale('log')
plt.yscale('log')

# Adding plot details
plt.title("Correlation Between Total CO2 Emissions and Average CO2 Per Capita (Log Scale)", fontsize=18)
plt.xlabel("Total CO2 Emissions (Kilotons, Log Scale)", fontsize=14)
plt.ylabel("Average CO2 Emissions Per Capita (Metric Tons, Log Scale)", fontsize=14)
plt.grid(True, which="both", linestyle='--', linewidth=0.7)

# Display the plot
plt.show()
```
